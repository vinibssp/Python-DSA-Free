## Pergunta de Negócio 1:

### Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

```python
df_pn1 = df_dsa[df_dsa['Categoria'] == 'Office Supplies']
```

```python
df_pn1.head()
```

```python
df_pn1_total = df_pn1.groupby('Cidade')['Valor_Venda'].sum()
print(df_pn1_total)
```

```python
maior_venda = df_pn1_total.idxmax()
print("A cidade de '{}' foi a cidade que teve o maior valor de venda para Office Supplies.".format(maior_venda))
```

## Pergunta de Negócio 2:

### Qual o Total de Vendas Por Data do Pedido?

Demonstre o resultado através de um gráfico de barras.

```python
df_pn2 = df_dsa.groupby('Data_Pedido')['Valor_Venda'].sum()
df_pn2.head()
```

```python
plt.figure(figsize = (15, 5))
df_pn2.plot(x = 'Data_Pedido', y = 'Valor_Venda')
plt.title('Total de Vendas Por Data do Pedido')
plt.show()
```

## Pergunta de Negócio 4:

### Quais São as 10 Cidades com Maior Total de Vendas?

Demonstre o resultado através de um gráfico de barras.

```python
df_pn4 = df_dsa.groupby('Cidade')['Valor_Venda'].sum().reset_index().sort_values(by = 'Valor_Venda', ascending = False).head(10)
df_pn4.head(10)
```

```python
plt.figure(figsize = (16, 6))
sns.barplot(data = df_pn4, 
            y = 'Valor_Venda', 
            x = 'Cidade').set(title = 'As 10 Cidades com Maior Total de Vendas')
plt.show()
```
